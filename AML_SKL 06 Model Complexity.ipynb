## Model Complexity: Overfitting and Underfitting
* Understanding overfitting and underfitting is key to constructing effective ML models
* Simple example using one-dimensional regression task

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Simple Toy Regression Example
* Underlying TRUE function in blue, noisy example from this function in green
* X axis is the input feature and the y axis is the outcome we want to predict
* We want to learn the true function using the Kneighbors regressor, predict mean of k nearest points

#### Number of Neigbors
* If number of neighbors is 2 (left panel), estimate (green) captures training set well, but focuses too much on noise in training set and is not likely to generalize well to new TEST data: OVERFITTING 
* If number of neighbors is 5 (center), we still capture most of true variation, but reduce some of the noise
* If number of neighbors is too high (right), we get rid of most of the noise, but our model is not flexible enough to express the variation that comes from the true function: UNDERFITTING

## Tradeoff between OVER and UNDER fitting
* From considering few neighbors to too many neighbors places restrictions on the function we can express. 
* These restrictions protect us from putting focusing too much on the particularities of the training set. 
* Adding too many restrictions can prevent us from 'learning' the true function of the task

### General ML Theme: Finding the "Sweetspot"
* In considering the relation between model complexity and accuracy or model perfomance in predictin new TEST data
* UNDERDITTING: If we choose a very restricted model (20 neighbors), we don't have flexibility to predict well; the model was not fitted well enough to the TRAIN set
* OVERFITTING: increasing the complexity of the model will allow us to better fit training data; however, if we make our model too flexible, it will fit the noise and particularities of TRAIn set, and will not generalize well to new TEST data 
* "Sweetspot" is found in the tradeoff between underfitting and overfitting, found by adjusting hyper-parameters
